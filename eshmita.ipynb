```python
import pandas as pd
import numpy as np
import gzip
from io import StringIO
```

```python
def load_geo_series_matrix(path):
    with gzip.open(path, 'rt') as f:
        lines = f.readlines()

    start, end = None, None
    for i, line in enumerate(lines):
        if line.startswith("!series_matrix_table_begin"):
            start = i + 1
        elif line.startswith("!series_matrix_table_end"):
            end = i
            break

    if start is None or end is None:
        raise ValueError("Series matrix table not found")

    data = ''.join(lines[start:end])
    df = pd.read_csv(StringIO(data), sep="\t", index_col=0)
    return df
```

```python
path = "/workspaces/Reasearch_Genomic/GSE72267_series_matrix.txt.gz"

df = load_geo_series_matrix(path)
print("Original shape:", df.shape)
```

```python
X = df.T
print("After transpose:", X.shape)
```

```python
y = np.zeros(X.shape[0])
y[: X.shape[0] // 2] = 1
```

```python
X = X.apply(pd.to_numeric, errors="coerce")
X = X.fillna(0)

print(X.dtypes.unique())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.linear_model import LogisticRegression

lasso = LogisticRegression(
    penalty="l1",
    solver="saga",
    max_iter=5000,
    C=1.0
)

lasso.fit(X_scaled, y)
```

```python
coef = lasso.coef_[0]
importance = np.abs(coef)
genes = X.columns.to_numpy()

lasso_df = pd.DataFrame({
    "Gene": genes,
    "Coefficient": coef,
    "Importance": importance
})
```

```python
top_32 = (
    lasso_df[lasso_df["Importance"] > 0]
    .sort_values(by="Importance", ascending=False)
    .head(32)
)

top_32
```

```python
top_32.to_csv("GSE72267_Top32_LASSO.csv", index=False)
print("Saved: GSE72267_Top32_LASSO.csv")
```

```python
selected_genes = top_32["Gene"].values
X_lasso = X[selected_genes]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(
    X_lasso,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, log_loss
```

```python
mlp = MLPClassifier(
    hidden_layer_sizes=(64, 32),
    activation="relu",
    solver="adam",
    batch_size=16,
    learning_rate_init=0.001,
    max_iter=1,
    warm_start=True,
    random_state=42
)
```

```python
epochs = 50

train_losses, val_losses = [], []
train_acc, val_acc = [], []

for _ in range(epochs):
    mlp.fit(X_train_scaled, y_train)

    y_train_pred = mlp.predict(X_train_scaled)
    y_val_pred = mlp.predict(X_val_scaled)

    y_train_prob = mlp.predict_proba(X_train_scaled)
    y_val_prob = mlp.predict_proba(X_val_scaled)

    train_acc.append(accuracy_score(y_train, y_train_pred))
    val_acc.append(accuracy_score(y_val, y_val_pred))

    train_losses.append(log_loss(y_train, y_train_prob))
    val_losses.append(log_loss(y_val, y_val_prob))
```

```python
print(f"Training Accuracy: {train_acc[-1]:.4f}")
print(f"Validation Accuracy: {val_acc[-1]:.4f}")
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(train_losses, label="Training Loss")
plt.plot(val_losses, label="Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("ANN Loss Curve (GSE72267)")
plt.legend()
plt.show()
```

```python
plt.figure()
plt.plot(train_acc, label="Training Accuracy")
plt.plot(val_acc, label="Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("ANN Accuracy Curve (GSE72267)")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import f1_score, classification_report

train_f1 = f1_score(y_train, y_train_pred)
val_f1 = f1_score(y_val, y_val_pred)

print(f"Train F1: {train_f1:.4f}")
print(f"Val F1:   {val_f1:.4f}")
print("\nValidation classification report:\n", classification_report(y_val, y_val_pred))
```

```python
from sklearn.metrics import recall_score

# Calculate recall for training set
train_recall_score = recall_score(y_train, y_train_pred)

# Calculate recall for validation set
val_recall_score = recall_score(y_val, y_val_pred)

print(f"Training Recall: {train_recall_score:.4f}")
print(f"Validation Recall: {val_recall_score:.4f}")
```

```python
import pickle

# Find the epoch with minimum validation loss
best_epoch = np.argmin(val_losses)
best_val_loss = val_losses[best_epoch]

print(f"Best epoch: {best_epoch + 1}")
print(f"Best validation loss: {best_val_loss:.6f}")

# Retrain the model up to the best epoch to get the best weights
mlp_best = MLPClassifier(
    hidden_layer_sizes=(64, 32),
    activation="relu",
    solver="adam",
    batch_size=16,
    learning_rate_init=0.001,
    max_iter=1,
    warm_start=True,
    random_state=42
)

for epoch in range(best_epoch + 1):
    mlp_best.fit(X_train_scaled, y_train)

# Save the best model
save_path = "mlp_best.pkl"
with open(save_path, "wb") as f:
    pickle.dump(mlp_best, f)

print(f"Best model saved to: {save_path}")

# Evaluate best model on validation set
y_val_pred_best = mlp_best.predict(X_val_scaled)
y_val_prob_best = mlp_best.predict_proba(X_val_scaled)
val_loss_best = log_loss(y_val, y_val_prob_best)
val_acc_best = accuracy_score(y_val, y_val_pred_best)
val_f1_best = f1_score(y_val, y_val_pred_best)
val_recall_best = recall_score(y_val, y_val_pred_best)

print(f"\nBest Model Validation Metrics:")
print(f"Accuracy: {val_acc_best:.4f}")
print(f"F1 Score: {val_f1_best:.4f}")
print(f"Recall: {val_recall_best:.4f}")
print(f"Loss: {val_loss_best:.6f}")
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Compute confusion matrix
cm = confusion_matrix(y_val, y_val_pred_best)

# Display the confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Class 0', 'Class 1'])
disp.plot(cmap='Blues')
plt.title("Confusion Matrix - Validation Set")
plt.show()
```
