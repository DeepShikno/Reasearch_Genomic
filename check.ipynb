```python
datasets = {
    "GSE6613": gse6613,
    "GSE72267": gse72267,
    "GSE99039": gse99039,
    "GSE57475": gse57475,
    "GSE18838": gse18838
}

for name, df in datasets.items():
    print(name, "Shape:", df.shape)
```

```python
for name, df in datasets.items():
    print(name)
    print(df.index[:5])
    print()
```

```python
common_genes = set(gse6613.index)

for df in datasets.values():
    common_genes = common_genes.intersection(set(df.index))

print("Number of common genes:", len(common_genes))
```
