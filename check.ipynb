```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np

scaled_datasets = {}

for name, df in datasets.items():
    
    print(f"\nProcessing {name}...")
    
    # 1️⃣ Check if dataframe is empty
    if df is None or df.empty:
        print(f"⚠ Skipping {name} — dataframe is empty")
        continue
    
    # 2️⃣ Ensure numeric type
    df = df.apply(pd.to_numeric, errors='coerce')
    
    # 3️⃣ Drop genes that became fully NaN
    df = df.dropna(axis=0, how='all')
    
    # 4️⃣ Fill remaining NaN with gene mean
    df = df.apply(lambda x: x.fillna(x.mean()), axis=1)
    
    # 5️⃣ Final safety check
    if df.shape[0] == 0:
        print(f"⚠ Skipping {name} — no genes remaining after cleaning")
        continue
    
    # 6️⃣ Scale per gene (standardize across samples)
    scaler = StandardScaler()
    scaled_array = scaler.fit_transform(df.T).T
    
    # 7️⃣ Rebuild dataframe
    scaled_df = pd.DataFrame(
        scaled_array,
        index=df.index,
        columns=df.columns
    )
    
    scaled_datasets[name] = scaled_df
    
    print(f"✅ {name} scaled successfully: {scaled_df.shape}")

print("\n🎯 Normalization complete.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))

for name, df in scaled_datasets.items():
    sns.kdeplot(df.values.flatten(), label=name)

plt.legend()
plt.title("Expression Distribution Comparison")
plt.show()
```

```python
import numpy as np

gene_means = {}

for name, df in scaled_datasets.items():
    gene_means[name] = df.mean(axis=1)

gene_means_df = pd.DataFrame(gene_means)

correlation_matrix = gene_means_df.corr()

print(correlation_matrix)
```

```python
from sklearn.decomposition import PCA

# Combine samples
combined = np.hstack([df.values for df in scaled_datasets.values()])

pca = PCA(n_components=2)
X_pca = pca.fit_transform(combined.T)

# Create dataset labels
labels = []
for name, df in scaled_datasets.items():
    labels += [name] * df.shape[1]

plt.figure(figsize=(8,6))

for dataset in scaled_datasets.keys():
    idx = [i for i,l in enumerate(labels) if l == dataset]
    plt.scatter(X_pca[idx,0], X_pca[idx,1], label=dataset)

plt.legend()
plt.title("PCA Batch Effect Check")
plt.show()
```
